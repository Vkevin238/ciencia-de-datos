```python
import pandas as pd
import requests 
import matplotlib.pyplot as plt 
api_url = 'https://www.datos.gov.co/resource/s7bv-9x6g.json'
response = requests.get(api_url)
if response.status_code == 200:
    data = response.json()

    df = pd.json_normalize(data)
    
 
else:
    print('no hay conexion')

cantidad = df.sort_values(by='total_poblaci_n_proyecci', ascending=False)
top_ten = cantidad.head(5)
plt.plot(top_ten['nombre_del_resguardo'],top_ten['total_poblaci_n_proyecci'])
```
